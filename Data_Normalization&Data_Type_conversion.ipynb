## A: Data Normalisation

```python
# Dataset
import pandas as pd
import numpy as np

data ={
    'Product' : ['Laptop','Mobile', 'Tablet', 'Headphones', 'Smartwatch', 'Camera'],
    'Price' : [55000, 20000, 30000, 2000, 8000, 45000],
    'Units_Sold' :  [25,60,35,80,50,15],
    'Discount' : [10,15,12,20,18,8]
}

df = pd.DataFrame(data)

print("Original Dataset")
print(df)
```

### Min - Max Normalization

```python
#1. Min - Max Normalization for Price

df['Price_MinMax'] = (df['Price'] - df['Price'].min()) / (df['Price'].max() - df['Price'].min())

print("\nAfter Min-Max Normalization")
print(df[['Product', 'Price', 'Price_MinMax']])
```

```python
# Min - Max Normalization for Units Sold

df['Units_Sold_MinMax'] = (df['Units_Sold'] - df['Units_Sold'].min()) / (df['Units_Sold'].max() - df['Units_Sold'].min())

print("\nAfter Min-Max Normalization")
print(df[['Product', 'Units_Sold', 'Units_Sold_MinMax']])
```

```python
# Min - Max Normalization for Discount

df['Discount_MinMax'] = (df['Discount'] - df['Discount'].min()) / (df['Discount'].max() - df['Discount'].min())

print("\nAfter Min-Max Normalization")
print(df[['Product', 'Discount', 'Discount_MinMax']])
```

### Z- Score Normalization

```python
#2. Z - Score Normalization

df['Units_Zscore'] = (df['Units_Sold'] - df['Units_Sold'].mean()) / df['Units_Sold'].std()

print("\nAfter Z-Score Normalization")
print(df[['Product', 'Units_Sold', 'Units_Zscore']])
```

### Decimal Scaling

```python
#3. Decimal  Scaling
# Decimal  Scaling for Price

df['Price_Decimal'] = df['Price'] / 100000

print("\nAfter Decimal Scaling")
print(df[['Product', 'Price', 'Price_Decimal']])
```

```python
# Decimal  Scaling for Discount

df['Discount_Zscore'] = df['Discount'] / 100

print("\nAfter Decimal Scaling")
print(df[['Product', 'Discount', 'Discount_Zscore']])
```

```python
# Normalize Multiple Columns

cols = ['Price', 'Units_Sold', 'Discount']

df_norm = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())

print("\nNormalised Dataset")
print(df_norm)
```

## B: Turning Quanlitative Variables into  Quantitative Variables in Python

```python
data =  {
    'Order_ID' : [101,102,103,104,105,106],
    'Customer_Gender' : ['Male', 'Female', 'Female', 'Male', 'Male', 'Female'],
    'Payment_Method' : ['UPI', 'Credit Card', 'Debit Card', 'COD', 'UPI', "Debit Card"],
    'Product_Category' :  ['Electronics', 'Clothing', 'Home', 'Electronics', 'Beauty','Electronics'],
    'City' : [ 'Pune', 'Mumbai', 'Delhi', 'Bangalore', 'Pune',  'Hyderabad'],
    'Order_Value' :  [4500, 2300, 3200, 5200, 1800, 4100]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Method : 1 - Label Encoding

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Gender_Label'] = le.fit_transform(df['Customer_Gender'])

print(df)
```

```python
# Method 2: One Hot Encoding

df_encoded = pd.get_dummies(df, columns=['Payment_Method'])

print(df_encoded)
```

```python
# Method 3: Encoding Multiple Columns

df_multi = pd.get_dummies(df, columns=['Product_Category', 'City'])

print(df_multi)
```

```python
# Method 4: Dummy Encoding (By Default it drops the first  column)

df_dummy = pd.get_dummies(df, columns = ['Payment_Method'], drop_first = True)

print(df_dummy)
```
